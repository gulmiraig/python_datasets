# About dataset

https://www.kaggle.com/datasets/ashpalsingh1525/netflix 

The Netflix dataset contains information on movies, including the release year, duration, cast, brief summaries, and ratings . It also includes additional information such as subtitles and audio languages and most importantly the mood experienced by the people while waching the movie. This data can be used for recommendation systems, data analysis, and research. The dataset includes over 550+ titles.

```python
plt.figure(figsize = (10,5))
df['release_year'].value_counts().sort_index().plot()

# This graph represents development of industry and how movie production changed over years

# Let's add x and y label 

plt.xlabel('Production Years', size = 14)
plt.ylabel('Produced movie numbers', size = 14)
plt.title('Movie production over years', size = 14)
```

# Let's do the same for maturity rating and see how everything changed over years

```python
df['maturity_rating'].value_counts().sort_index()
```

```python
df[['release_year', 'maturity_rating']].value_counts().sort_index()
```

```python
df[['release_year', 'maturity_rating']].value_counts().sort_index().unstack()
```

```python
# get frequency count based on 2 columns 
```

```python
plt.rcParams["figure.figsize"] = (15, 8)
df[['release_year', 'maturity_rating']].value_counts().sort_index().unstack().plot(kind = 'bar'); 



plt.xlabel('Release year', size = 14)
plt.ylabel('Produced movie numbers', size = 14)
plt.title('Movie production over years based on maturity rating', size = 14);
```

#  <span style='color: Teal'> We can conclude that over years movie production is increasing accross all types of maturity rating except U type. </span>

# Checking my dataset

```python
df.head()
```

```python
plt.figure(figsize = (10,4))
sns.countplot(x=df['maturity_rating'], palette = 'summer')
plt.title('Produced movies based on the maturity rating', size = 14);

sns.despine()
```

```python
plt.figure(figsize = (20,10))
sns.countplot(x=df['mood'], palette = 'viridis')

plt.xticks(rotation=45, ha="right", fontsize=14)  # horizontalalignment or  ha


plt.xlabel('Mood', size = 18)
plt.ylabel('Produced movies count', size = 18)
plt.title('Produced movies based on the mood type', size = 20);
sns.despine()
```
